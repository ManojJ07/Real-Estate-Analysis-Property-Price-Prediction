```python
location = df['location'].tolist()

print(location)
```

```python
pattern = r"in\s(.*)"

#list to store matching strings
loc = []

for string in location:
    match = re.search(pattern, string, re.IGNORECASE)
    if match:
        loc.append(match.group(1))

print(loc)
```

```python
len(loc)
```

```python
pattern = r"(.*?)\sin"

#list to store modified strings
modified_strings = []

for string in location:
    match = re.search(pattern, string, re.IGNORECASE)
    if match:
        modified_strings.append(match.group(1))
    else:
        modified_strings.append(string)

print(modified_strings)
```

```python
words_to_drop = ["BHK"]

#regular expression pattern to match and drop words
pattern = "|".join(words_to_drop)

#list to store modified strings
modified_strings1 = []

for string in modified_strings:
    modified_string = re.sub(pattern, "", string, flags=re.IGNORECASE)
    modified_strings1.append(modified_string)

print(modified_strings1)
```

```python
pattern = "[0-9]+"

#list to store modified strings
modified_strings2 = []

for string in modified_strings1:
    modified_string = re.sub(pattern, "", string)
    modified_strings2.append(modified_string)

print(modified_strings2)
```

```python
#regular expression pattern to match and remove space before starting word
pattern = "^\s"

#list to store modified strings
land = []

for string in modified_strings2:
    modified_string = re.sub(pattern, "", string)
    land.append(modified_string)

print(land)
```

```python
#regular expression pattern to match and remove space before starting word
pattern = "^\s"

#list to store modified strings
type_of_land = []

for string in land:
    modified_string = re.sub(pattern, "", string)
    type_of_land.append(modified_string)

print(type_of_land)
```

```python
from collections import Counter

# Count the occurrences of each string in the list
counts = Counter(type_of_land)

# Get the most common string
most_common = counts.most_common(1)[0][0]

# Replace empty strings with the most common string
type_of_land = [most_common if x=='' else x for x in type_of_land]
```

```python
type_of_land = list(map(lambda x: 'Bedroom House' if x=='House' else x, type_of_land))
```

# price per sqft

```python
price = df['price_per_sqft'].tolist()

print(price)
```

```python
characters_to_remove = ["₹ ","/sq.ft.",","]

#regular expression pattern to match and remove special characters and characters
pattern = "[{}]".format("".join(characters_to_remove))

#list to store modified strings
Price_per_sqft = []

for string in price:
    modified_string = re.sub(pattern, "", string)
    Price_per_sqft.append(modified_string)

print(Price_per_sqft)
```

```python
Price_per_SqFt = [int(x) for x in Price_per_sqft]

print(Price_per_SqFt)
```

```python
characters_to_remove = ["sq.ft."," sq.m.) Built-up Area",","]

#regular expression pattern to match and remove special characters and characters
pattern = "[{}]".format("".join(characters_to_remove))

#list to store modified strings
total = []

for string in sqft:
    modified_string = re.sub(pattern, "", string)
    total.append(modified_string)

print(total)
```

```python
Total_SqFt = [int(x) for x in total_sqft]

print(Total_SqFt)
```

# Total price

```python
total_price = df['price'].tolist()

print(total_price)
```

```python
for price in total_price:
    match = re.search(r'₹ (.*)', price)
    print(match.group(1))
```

```python
for price in total_price:
    matches = re.findall(r'₹\s*\d+[.,]?\d*', price)
    print(matches)
```

```python
matches = []

for price in total_price:
    matches.extend(re.findall(r'₹\s*\d+(?:[,.]\d+)*', price))

print(matches)
```

```python
new_matches = matches[0::2]
print(new_matches)
```

```python
new_matches = [str(x.replace("₹ ","")) if type(x) == str else x for x in new_matches]
print(new_matches)
```

```python
for i, val in enumerate(new_matches):
    if ',' in val:
        # keep as it is
        continue
    elif '.' in val:
        if val.index('.') == 1:
            new_matches[i] = float(val) * 10000000
        else:
            new_matches[i] = float(val) * 100000
    elif len(val) == 2:
        new_matches[i] = int(val) * 100000
    else:
        new_matches[i] = int(val) * 10000000
```

```python
for i, val in enumerate(new_matches):
    if isinstance(val, str):
        val = val.replace(',','') # remove the comma from the string
    new_matches[i] = int(val) # convert to integer
```
