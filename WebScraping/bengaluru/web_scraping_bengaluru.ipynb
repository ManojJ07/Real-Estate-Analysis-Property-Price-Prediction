# selenium

```python
import selenium
import pandas as pd
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
import time
from bs4 import BeautifulSoup 
```

```python
path = "C:/Users/imjma/chromedriver.exe"
url = "https://www.99acres.com/search/property/buy/bangalore-all?city=20&preference=S&area_unit=1&budget_min=0&res_com=R&isPreLeased=N"
```

```python
location = []
price_per_sqft = []
total_sqft = []
built_area_sqm =[]
num_of_bhk = []
price = []
```

```python
def scroll_down():

    y = 750
    
    while True:

        # Scroll down to the bottom.
        driver.execute_script("window.scrollTo(0, " + str(y) + ")")
        # to get the document
        x = driver.execute_script("return document.documentElement.scrollHeight")
        print(f"x = {x}")
        print(f"y = {y}")
        y += 750
        time.sleep(2)
        
        
       
        

        if y > x:
            html=driver.page_source
            soup = BeautifulSoup(html,'html.parser')
            data = soup.find_all('section')
           
            for i in data:
                
    
                location.append(i.find('h2', class_="srpTuple__tupleTitleOverflow"))
    
                price_per_sqft.append(i.find('div', id="srp_tuple_price_per_unit_area"))
    
                total_sqft.append(i.find('td', id="srp_tuple_primary_area"))

                built_area_sqm.append(i.find('div', id="srp_tuple_secondary_area"))

                num_of_bhk.append(i.find('td', id="srp_tuple_bedroom"))
    
                price.append(i.find('td', id="srp_tuple_price"))
    
    


            
            element = driver.find_element(By.CLASS_NAME, "Pagination__srpPagination")
            driver.execute_script("window.scrollTo(0, " + str(element.location['y'] - 300) + ")") 
            time.sleep(2)
            WebDriverWait(driver, 20).until(EC.element_to_be_clickable((By.XPATH, "//a[@class='list_header_bold' and contains(., 'Next Page ')]"))).click()
            
            break
```

```python
chrome_options = Options()
chrome_options.add_argument("--start-maximized")
driver = webdriver.Chrome(path, chrome_options=chrome_options)
driver.get(url)

for i in range(25):
    scroll_down()
```
